```python
#시각화를 위한 라이브러리 불러오기
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore") #오류무시
```

```python
#데이터 불러오기
df=pd.read_csv("customer data.csv",encoding="ISO-8859-1")
```

```python
df
```

```python
#NAN 값들이 들어있는 행 지우기
df.dropna(subset=['CustomerID'],how='all',inplace=True)
df
```

```python
#가장 최근 구매날짜
df['InvoiceDate'].max()
```

```python
#현재날짜를 2011.12.9일로 지정
import datetime as dt
now = dt.date(2011,12,9)
```

```python
df['date'] = pd.DatetimeIndex(df.InvoiceDate).date
df
```

```python
#고객이 구매한 마지막 날짜까지 고객별로 그룹화

recency_df = df.groupby(['CustomerID'],as_index=False)['date'].max()
recency_df.columns = ['CustomerID','LastPurchaseDate']
```

```python
#며칠간의 최신 날짜를 참조하여 얼마나 자주 구매하는지 계산
recency_df['Recency'] = recency_df.LastPurchaseDate.apply(lambda x : (now - x).days)
recency_df.head() #df에서 위에 5개의 항목만 보기
```

```python
#최근 구매횟수를 모아둔 Recency라는 열을 만들었으므로, 기존의 날짜 데이터 열은 지운다.
recency_df.drop(columns=['LastPurchaseDate'],inplace=True)
recency_df
```

```python
#고객들의 거래횟수(빈도) 알아보기
frequency_df = df.copy()
frequency_df.drop_duplicates(subset=['CustomerID','InvoiceNo'], keep="first", inplace=True)
frequency_df = frequency_df.groupby('CustomerID',as_index=False)['InvoiceNo'].count()
frequency_df.columns = ['CustomerID','Frequency']
frequency_df.head()
```

```python
#고객이 각 거래에서 얼마나 소비하는지 계산
df['Total_cost'] = df['UnitPrice'] * df['Quantity']
#최근 날짜에 대한 고객의 지출 내역을 종합 검사
monetary_df=df.groupby('CustomerID',as_index=False)['Total_cost'].sum()
monetary_df.columns = ['CustomerID','Monetary']
monetary_df.head() #새로 만든 monetary 열에 대한 data 위에서 5개만 출력
```

```python
monetary_df.sort_values(by=['Monetary'], ascending=[True])
```

```python
#RFM을 attribute로 갖는 data frame 재구성
rf = recency_df.merge(frequency_df,left_on='CustomerID',right_on='CustomerID')
rfm = rf.merge(monetary_df,left_on='CustomerID',right_on='CustomerID')
rfm.set_index('CustomerID',inplace=True)
rfm
```

```python
rfm_segmentation = rfm.copy() #사분위수를 단일 데이터로 가져옴

#K-Means함수를 이용하여 clustering
from sklearn.cluster import KMeans

#Elbow method 기법을 사용하여 적절한 k의 수를 찾는다.
#클러스터 수 1~20개를 루프
Nc = range(1, 20)
kmeans = [KMeans(n_clusters=i) for i in Nc]
score = [kmeans[i].fit(rfm_segmentation).score(rfm_segmentation) for i in range(len(kmeans))]
plt.plot(Nc,score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

```python
#Elbow Cuve 그래프를 통해 k=3이 적절함을 알 수 있으므로
#k=3으로 클러스터링
k = KMeans(n_clusters=3).fit(rfm_segmentation) #학습
print(k.n_clusters)
```

```python
rfm_segmentation['cluster'] = k.labels_
rfm_segmentation
```

```python
#R의 5분위수(Rscore)
'''
Recency는 값이 작을수록 score가 좋은 것이므로,

0%~20%:1사분위, 20%~40%:2사분위,
40%~60%: 3사분위, 60%~80%: 4사분위, 80%~100%: 5사분위
'''
#인수(x = 값, p = Recency, d = 사분위수 지시)
def RScore(x,p,d):
    if x <= d[p][0.20]:
        return 5
    elif x <= d[p][0.40]:
        return 4
    elif x <= d[p][0.60]: 
        return 3
    elif x <= d[p][0.80]: 
        return 2
    else:
        return 1
```

```python
#FM의 5분위수(Rscore)
'''
Frequency와 Monetary는 값이 클수록 score가 좋은 것이므로,

0%~20%:5사분위, 20%~40%:4사분위,
40%~60%: 3사분위, 60%~80%: 2사분위, 80%~100%: 1사분위
'''
#인수(x = 값, p = Frequency와 Monetary, d = 사분위수 지시)
def FMScore(x,p,d):
    if x <= d[p][0.20]:
        return 1
    elif x <= d[p][0.40]:
        return 2
    elif x <= d[p][0.60]: 
        return 3
    elif x <= d[p][0.80]: 
        return 4
    else:
        return 5
```

```python
#5분위수 값들의 레벨은 20% 40% 60% 80% 100%와 같다.
#5사분위수의 경계 값을 알기 위함
'''
예를 들어 1사분위수는 0%~20%에 100개의 값이 있는 경우이고, 2사분위는 다음 20%를 차지하는 20%~40%에 있는 값들이고, 
3사분위는 다음 20%를 차지하는 40%~60%에 있는 값들이고,4사분위는 다음 20%를 차지하는 60%~80%에 있는 값들이고,
5사분위수는 다음 20%에 60%~100%에 있는 값들이다.
'''

quantile = rfm.quantile(q=[0.2,0.4,0.6, 0.8])
quantile
```

```python
#각각의 고객들의 Recency, Frequency, Monetary가 5사분위수 중 어느 등급에 해당하는지 알아보기 위한 열 추가
rfm_segmentation['R_Quartile'] = rfm_segmentation['Recency'].apply(RScore,args=('Recency',quantile))
rfm_segmentation['F_Quartile'] = rfm_segmentation['Frequency'].apply(FMScore, args=('Frequency',quantile))
rfm_segmentation['M_Quartile'] = rfm_segmentation['Monetary'].apply(FMScore, args=('Monetary',quantile))
```

```python
#각각의 고객들의 Recency, Frequency, Montary의 값들을 고려하여 RFM score을 계산한 열 추가(등급고려 X)
rfm_segmentation['RFMScore'] = rfm_segmentation.R_Quartile.map(str) \
                                    + rfm_segmentation.F_Quartile.map(str) \
                                    + rfm_segmentation.M_Quartile.map(str)
rfm_segmentation
```

```python
#각각의 고객들의 Recency, Frequency, Montary의 등급을 고려하여 RFM의 최종 Total score를 계산한 열 추가
#Total score는 3~12 사이의 값들을 갖게 됨
rfm_segmentation['Total_score'] = rfm_segmentation['R_Quartile'] + rfm_segmentation['F_Quartile'] + rfm_segmentation['M_Quartile']
rfm_segmentation
```
